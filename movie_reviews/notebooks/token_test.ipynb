```python
import os
import openai
from dotenv import load_dotenv

# Set up Azure OpenAI
openai.api_type = "azure"
openai.api_base = os.getenv("SOUTH_CENTRAL_US_OPENAI_API_BASE")
openai.api_version = "2022-12-01"
openai.api_key = os.getenv("SOUTH_CENTRAL_US_OPENAI_API_KEY")
```

```python
import pandas as pd

df = pd.read_csv("../data/rottentomatoes-20movies-wordcount.csv", sep='\t')
```

```python
import tiktoken
tokenizer = tiktoken.get_encoding("cl100k_base")

input = 'Movie title: ' + df['Movie'][0] + '\n' + df['Review'][0]

sample_encode = tokenizer.encode(input) 
decode = tokenizer.decode_tokens_bytes(sample_encode)

len(decode)
```

```python
print(len(df))
```

```python
token_len = 0
for i in range(len(df)):
    input = 'Movie title: ' + df['Movie'][i] + '\n' + df['Review'][i]

    encode = tokenizer.encode(input)
    decode = tokenizer.decode_tokens_bytes(encode)

    token_len = token_len + len(decode)
    print(token_len)
```
